```python
scan1_re = s1r1r2r3[(s1r1r2r3['ViewingCondition'] == 'reappraisal') & (s1r1r2r3['ImageCategory'] == 'negative')]
scan2_re = s2r1r2r3[(s2r1r2r3['ViewingCondition'] == 'reappraisal') & (s2r1r2r3['ImageCategory'] == 'negative')]
scan3_re = s3r1r2r3[(s3r1r2r3['ViewingCondition'] == 'reappraisal') & (s3r1r2r3['ImageCategory'] == 'negative')]

scan1_pa = s1r1r2r3[(s1r1r2r3['ViewingCondition'] == 'passive') & (s1r1r2r3['ImageCategory'] == 'negative')]
scan2_pa = s2r1r2r3[(s2r1r2r3['ViewingCondition'] == 'passive') & (s2r1r2r3['ImageCategory'] == 'negative')]
scan3_pa = s3r1r2r3[(s3r1r2r3['ViewingCondition'] == 'passive') & (s3r1r2r3['ImageCategory'] == 'negative')]


scan1_re_dying = s1r1r2r3[(s1r1r2r3['ViewingCondition'] == 'reappraisal') & (s1r1r2r3['ImageCategory'] == 'negative') & (s1r1r2r3['ImageSubcategory'] == 'dying')]
scan2_re_dying = s2r1r2r3[(s2r1r2r3['ViewingCondition'] == 'reappraisal') & (s2r1r2r3['ImageCategory'] == 'negative') & (s2r1r2r3['ImageSubcategory'] == 'dying')]
scan3_re_dying = s3r1r2r3[(s3r1r2r3['ViewingCondition'] == 'reappraisal') & (s3r1r2r3['ImageCategory'] == 'negative') & (s3r1r2r3['ImageSubcategory'] == 'dying')]

scan1_pa_dying = s1r1r2r3[(s1r1r2r3['ViewingCondition'] == 'passive') & (s1r1r2r3['ImageCategory'] == 'negative') & (s1r1r2r3['ImageSubcategory'] == 'dying')]
scan2_pa_dying = s2r1r2r3[(s2r1r2r3['ViewingCondition'] == 'passive') & (s2r1r2r3['ImageCategory'] == 'negative') & (s2r1r2r3['ImageSubcategory'] == 'dying')]
scan3_pa_dying = s3r1r2r3[(s3r1r2r3['ViewingCondition'] == 'passive') & (s3r1r2r3['ImageCategory'] == 'negative') & (s3r1r2r3['ImageSubcategory'] == 'dying')]


scan1_re_general = s1r1r2r3[(s1r1r2r3['ViewingCondition'] == 'reappraisal') & (s1r1r2r3['ImageCategory'] == 'negative') & (s1r1r2r3['ImageSubcategory'] == 'general')]
scan2_re_general = s2r1r2r3[(s2r1r2r3['ViewingCondition'] == 'reappraisal') & (s2r1r2r3['ImageCategory'] == 'negative') & (s2r1r2r3['ImageSubcategory'] == 'general')]
scan3_re_general = s3r1r2r3[(s3r1r2r3['ViewingCondition'] == 'reappraisal') & (s3r1r2r3['ImageCategory'] == 'negative') & (s3r1r2r3['ImageSubcategory'] == 'general')]

scan1_pa_general = s1r1r2r3[(s1r1r2r3['ViewingCondition'] == 'passive') & (s1r1r2r3['ImageCategory'] == 'negative') & (s1r1r2r3['ImageSubcategory'] == 'general')]
scan2_pa_general = s2r1r2r3[(s2r1r2r3['ViewingCondition'] == 'passive') & (s2r1r2r3['ImageCategory'] == 'negative') & (s2r1r2r3['ImageSubcategory'] == 'general')]
scan3_pa_general = s3r1r2r3[(s3r1r2r3['ViewingCondition'] == 'passive') & (s3r1r2r3['ImageCategory'] == 'negative') & (s3r1r2r3['ImageSubcategory'] == 'general')]
```

```python
kr = 'KeyResponse'
```

```python
count_all_scan1 = scan1_pa[kr].count()
ERI_scan1 = scan1_re[kr].mean() - scan1_pa[kr].mean()
percent_change_ERI_scan1 = ERI_scan1 / scan1_pa[kr].mean()

ERI_dying_scan1 = scan1_re_dying[kr].mean() - scan1_pa_dying[kr].mean()
percent_change_ERI_dying_scan1 = ERI_dying_scan1 / scan1_pa_dying[kr].mean()

ERI_general_scan1 = scan1_re_general[kr].mean() - scan1_pa_general[kr].mean()
percent_change_ERI_general_scan1 = ERI_general_scan1 / scan1_pa_general[kr].mean()


count_all_scan2 = scan2_pa[kr].count()
ERI_scan2 = scan2_re[kr].mean() - scan2_pa[kr].mean()
percent_change_ERI_scan2 = ERI_scan2 / scan2_pa[kr].mean()

ERI_dying_scan2 = scan2_re_dying[kr].mean() - scan2_pa_dying[kr].mean()
percent_change_ERI_dying_scan2 = ERI_dying_scan2 / scan2_pa_dying[kr].mean()

ERI_general_scan2 = scan2_re_general[kr].mean() - scan2_pa_general[kr].mean()
percent_change_ERI_general_scan2 = ERI_general_scan2 / scan2_pa_general[kr].mean()


count_all_scan3 = scan3_pa[kr].count()
ERI_scan3 = scan3_re[kr].mean() - scan3_pa[kr].mean()
percent_change_ERI_scan3 = ERI_scan3 / scan3_pa[kr].mean()

ERI_dying_scan3 = scan3_re_dying[kr].mean() - scan3_pa_dying[kr].mean()
percent_change_ERI_dying_scan3 = ERI_dying_scan3 / scan3_pa_dying[kr].mean()

ERI_general_scan3 = scan3_re_general[kr].mean() - scan3_pa_general[kr].mean()
percent_change_ERI_general_scan3 = ERI_general_scan3 / scan3_pa_general[kr].mean()
```
